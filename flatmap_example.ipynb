# SurfAnalysis - example
# Basic Mapping and plotting
This is a example of how to map volumes to the surface and to plot the result on a flat map using the the SurfAnalysisPy toolbox. 

```python
# Import the necesarry modules for this example 
import surfAnalysisPy as surf
import numpy as np
import nibabel as nb
import os
```

```python
# Define the file names with paths used in this example
s02white = os.path.join('example_individual','s02.L.white.32k.surf.gii')
s02pial = os.path.join('example_individual','s02.L.pial.32k.surf.gii')
s02func = os.path.join('example_individual','con_TOM-average_4.nii')
s02sulc = os.path.join('example_individual','s02.L.sulc.32k.shape.gii')
flat = os.path.join('standard_mesh','fs_L','fs_LR.32k.L.flat.surf.gii')
```

## Mapping of volumes to surface
Description

```python
# Map a Nifti to the surface 
D = surf.map.vol_to_surf([s02func],s02pial,s02white)
```
